```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
X, y = make_regression(n_features=1, noise= 30, random_state= 0)
#Random state es una semilla
```

```python
X.shape, y.shape
```

```python
df = pd.DataFrame(X)
df.columns = ["x"]
df["y"]= y
df
```

```python
df['x'].plot(kind = "hist")
```

```python
y_mean = np.array([y.mean()] * 100) #Un valor de la media para cada valor en X
```

```python
#Selector
plt.figure(figsize=(10,3))

#Data
plt.scatter(X,y, label = "Random Data")

#Simple mean
plt.plot(X,y_mean, label = "Mean", c="green")

#Show
plt.xlabel("Features (x)")
plt.ylabel("Target (y)")
plt.title("Scatter Example Plot")
plt.legend()

#Save
plt.savefig("../reports/figures/random1.png")
```

```python
#Clase: lo que define el comportamiento de un modelo
#Objeto: la instancia de una clase -> objeto model del tipo LinearRegression
from sklearn.linear_model import LinearRegression


# Create an instance of the LinearRegression model
model = LinearRegression()

# Fit the model to your data
model.fit(X, y)
```

```python
y_pred= model.predict(X)
```

```python
#Selector
plt.figure(figsize=(10,3))

#Data
plt.scatter(X,y, label = "Random Data")

#Simple mean
plt.plot(X,y_pred, label = "Regression", c="red")
plt.plot(X,y_mean, label = "Mean", c="green")

#Show
plt.xlabel("Features (x)")
plt.ylabel("Target (y)")
plt.title("Scatter Example Plot")
plt.legend()

#Save
plt.savefig("../reports/figures/random1.png")
```
